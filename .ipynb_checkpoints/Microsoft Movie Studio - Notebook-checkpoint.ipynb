# Microsoft Movie Studio: Data-Driven Research

Microsoft has decided to launch a production studio developing original movies for the big screen. This analysis aims to provide insights regarding what type of film they should explore first. 

### The Data

The data used here comes from some of the most well-known and reliable sources in the industry, such as IMDB and Box Office Mojo, IMDB Pro's box office portal. Put together, these datasets contain information about thousands of movies, their box office performance domestically, and worldwide and rating score by IMDB users. Also, a dataset scraped from The Numbers contains production budget information for almost 6,000 successful movies. 

````python
#Importing all necessary libraries for EDA and visualization 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

### Importing Data

The first step is importing the four datasets and previewing their contents with the ```.head()``` method in Pandas.

````python
#IMDB TITLE BASICS --> title id, primary title, original title, start year, runtime, genres 
imdb_basics = pd.read_csv('data/imdb.title.basics.csv')

imdb_basics.head(3)
````

````python
#IMDB Title Ratings --> title id, average rating, number of votes
imdb_ratings = pd.read_csv('data/imdb.title.ratings.csv')

imdb_ratings.head(3)
````

````python
#The Numbers: Movie Budgets --> id, date, movie, production budget, dom_gross, ww_gross
movie_financials = pd.read_csv('data/tn.movie_budgets.csv')

movie_financials.head(3)
````

````python
#Box Office Mojo: title, studio, dom_gross, intl_gross, year
box_office = pd.read_csv('data/bom.movie_gross.csv')

box_office.head(3)
````

### Data Preparation

In this next step, the data from multiple sources will be brought together by merging the imported DataFrames. Some Series will need to be converted into a different format more suitable for plotting to further clean and prepare the data. 

Due to the movie industry's fast-changing nature, it makes the most sense to look at data about more recent releases. Therefore, the newly created DataFrames will be filtered to contain movies released in the last five years.  

````python
#Merge the two IMDB DataFrames into one on the title ID as the key:

imdb_merged = pd.merge(imdb_basics, imdb_ratings ,on='tconst')
````

````python
#Create a DataFrame for the movies in the IMDB database released in the last five years (2015 or later):

imdb_recent = imdb_merged.loc[imdb_merged['start_year'] >= 2015].drop(columns = ['runtime_minutes', 'original_title'])
````

The newly created imdb_recent DataFrame has 35,634 entries brokend down by year of release as follows:

2016  -  8721 movies

2017  -  8713 movies

2015  -  8494 movies

2018  -  7526 movies

2019  -  2180 movies

````python
#Rename the columns to ensure consistancy with other tables that will get merged later on.
imdb_recent.rename(columns={'primary_title':'movie',
                            'start_year':'year',
                            'averagerating': 'rating',
                            'numvotes': 'number of votes'},
                            inplace=True)
````

````python
#Preview of the newly created DataFrame
imdb_recent.head(3)
````

````python
#Exploring the data types of each Series 
imdb_recent.info()
````



One issue that becomes evident is that a movie can be classified under multiple genres, and the DataFrame lists all genres as comma-separated values in the 'genre' Series.  The order is alphabetical, and no genre appears as primary or secondary. Thus, each movie represents a unique instance of each genre.

````python
#Split the str of listed genres on the comma

imdb_recent['genres'] = imdb_recent['genres'].str.split(',')

#Break down the genres column into separate rows where each movie is an instance of all genres it belongs to

imdb_recent = (imdb_recent
 .set_index(['movie','year',"rating","number of votes"])['genres']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_4', axis=1)
 .rename(columns={0:'genres'}))
````

````python
#Check results
imdb_recent.head(5)
````

Next, the dataset from The Numbers contains information for movies from over ten years ago, and the release dates are specified by day, month, and year as opposed to just year of release. This makes it hard to filter out movies that were not released in the last five years.

All financial columns also contain their values as strings, including special characters such as $ and commas. These Series need to be converted to a numeric datatype for ease of plotting later. 

````python
#Convert the release date column of type 'obj' to a year column in the 'movie_financials' DataFrame
release_years = []
for year in movie_financials.release_date:
    release_years.append(year[-4:])
movie_financials['year'] = release_years
movie_financials['year'] = movie_financials['year'].astype(int)
````

````python
#Getting the financial data for the movies released between 2015 and 2019(incl):

recent_financials = movie_financials.loc[(movie_financials['year'] >= 2015) & (movie_financials['year'] < 2020) ].drop(columns = ['id', 'release_date'])
recent_financials.info()
````

The recent_financials DataFrame contains data for 935 movies broken down by year as follows:

2015 -  338 movies

2016 -  219 movies

2017 - 168 movies

2018 - 143 movies

2019 - 67 movies

````python
#Convert the $ amounts for production, domestic gross, and worldwide gross from str to int

def replace_dollars(x:str):
    return int(x.replace('$','').replace(',','').replace('.00',''))

#recent_financials[['production_budget','domestic_gross','worldwide_gross']] = recent_financials[['production_budget','domestic_gross','worldwide_gross']].apply(replace_dollars)

recent_financials['production_budget'] = recent_financials['production_budget'].apply(replace_dollars)
recent_financials['domestic_gross'] = recent_financials['domestic_gross'].apply(replace_dollars)
recent_financials['worldwide_gross'] = recent_financials['worldwide_gross'].apply(replace_dollars)
````

It will be helpful to have production budgets in the same DataFrame as the IMDB-verified box office data for domesticd and worldwide gross revenue. Merging the ```'box office'``` and the ```'recent financials'``` DataFrames will comibne that data. 

````python
#Change the name of the 'title' column to 'movie' so the column can be used as a key for the merger 
box_office['movie'] = box_office['title']
````

````python
studios = pd.merge(box_office, recent_financials ,on='movie')
````

````python
#Removing duplicate and irreelvant columns
revenue_df = studios.drop(columns = ['title', 'domestic_gross_x', 'foreign_gross', 'year_x'])
````

````python
revenue_df.head(3)
````

````python
#Check for duplicates
check_df = revenue_df[revenue_df['worldwide_gross'] == 0]
check_df.shape
````

````python
#Rename columns in the newly merged DataFrame
revenue_df.rename(columns={'production_budget':'budget', 
                          'domestic_gross_y':'domestic gross', 
                          'worldwide_gross':'ww gross',
                          'year_y':'year'},
                 inplace=True)
````

````python
revenue_df.head(3)
````

For ease of plotting, it will help to add two additional columns with calculcated ROI and international gross reveneue. Also, in order to explore the relationship between genre and revenue, one last DataFrame is created merging the revenue data with genre information from ```imdb_recent```.

````python
#ADD ROI column to studio financials:
revenue_df['roi'] = round(((revenue_df['ww gross']/revenue_df['budget'])*100), 2)
````

````python
#ADD International Revenue Column = WW Gross - Domestic Gross
revenue_df['intl gross'] = revenue_df['ww gross'] - revenue_df['domestic gross']
````

````python
#Merge genres into the Revenue DataFrame
revenue_genre = pd.merge(revenue_df, imdb_recent, on='movie')

#Remove columns that won't be used for analysis and plots 
revenue_genre.drop(columns = ['year_y', 'number of votes'], inplace=True)

#Rename year_x column
revenue_genre.rename(columns={'year_x':'year'}, inplace=True)
````

## Finalized DataFrames

## ANALYSIS

To help Microsoft decide what type of movies they should look into producing first, genre performance of recently released movies will be analyzed alongside user ratings and revenue at the box office. 

Thus, this analysis is focused on the three questions below: 

#### 1. What genres of movies have received the most favorable ratings on IMDB?

#### 2. What genres of movies have shown the highest ROI in the last five years?

#### 3. What genres of movies do the leading 10 production studios produce the most? 


### 1. What genres of movies have received the most favorable ratings on IMDB?

With over 83 million users, IMDB is the most popular website for movie lovers and one of the most frequently used platforms where they express their opinion on new and old films. Users do so by writing reviews or assigning a user rating on a scale from 1-10. For the purpose of this analysis, audience reception will be measured by the IMDB average user rating. 

The genres that accumulated favorable ratings (7 or above) are likely to be best received and most appreciated by movie-goers. 

````python
#Which 15 genres had the most movies released in the last fives years?
top_genres_df =imdb_recent.groupby('genres').size().sort_values(ascending=False).head(15)
````

Genre classificaiton can get very graniular leading to very cluttered plots. To avoid overplotting, the top 15 genres (by count of movies released) will be devided in three subsets and their average rating performance will be represented with a histogram.

````python
#Divide genres in three subsets for plotting
print(top_genres_df)
first_five = list(top_genres_df.index[:5])
next_five = list(top_genres_df.index[5:10])
final_five = list(top_genres_df.index[10:15])
````

````python
#Create a DataFrame for each subset
first_five_df = imdb_recent[imdb_recent['genres'].isin(first_five)]
next_five_df = imdb_recent[imdb_recent['genres'].isin(next_five)]
final_five_df = imdb_recent[imdb_recent['genres'].isin(final_five)]
````

````python
#Create a histogram for the first subset using Seaborn visualiztion 
ax_one = sns.histplot(data = first_five_df, x = 'rating', hue = 'genres', multiple="stack", palette='pastel')
ax_one.set_title("Top 15 Genres (1-5)", size = 16)
ax_one.set_xlabel('Average User Rating', size=13)
ax_one.set_ylabel('Count of Movies', size = 13)
sns.set(rc={'figure.figsize':(16,8)})
sns.set_style()
````

````python
#Create a histogram for the second subset 
ax_two = sns.histplot(data = next_five_df, x = 'rating', hue = 'genres', multiple="stack", palette='pastel')
ax_two.set_title("Top 15 Genres (6-10)", size = 16)
ax_two.set_xlabel('Average User Rating', size=13)
ax_two.set_ylabel('Count of Movies', size = 13)
sns.set(rc={'figure.figsize':(16,8)})
sns.set_style()
````

````python
#Create a histogram for the last subset 
ax_three = sns.histplot(data = final_five_df, x = 'rating', hue = 'genres', multiple="stack", palette='pastel')
ax_three.set_title("Top 15 Genres (11-15)", size = 16)
ax_three.set_xlabel('Average User Rating', size=13)
ax_three.set_ylabel('Count of Movies', size = 13)
sns.set(rc={'figure.figsize':(16,8)})
sns.set_style()
````

### Conclusion

#### Summary:

The vast majority of **Documentaries** receive an above-average rating (6+) by users, with a good amount of movies being rated above 8 as well. 

Many **Horror** movies receive poor, below-average ratings, with very few of them receiving a rating higher than 8.

**Thrillers** follow a similar trend with a slightly higher count of movies receiving average ratings.

**Biography** movies tend to receive most of their ratings in the 6.5-9 range.

**Romance** films' rating is pretty evenly distributed around the 6.5 ratings as the most frequently assigned user score. 

**History** movies receive predominantly ratings above 7 and almost no below-average ratings.

**Sci-Fi** seems like the most polarizing genre, with ratings ranging from 2 to slightly below 9. 

**Fantasy** follows almost a similar trend of even distribution, only starting closer to 3. 


#### Key Takeaways:

- If a movie's goal is to receive raving reviews from the audience, producing a fiction movie that includes biographical or historical components is a pretty safe way to get you there. Alternatively, making a documentary also has higher than average chances to receive favorable ratings from movie-goers. 
 
 
- If audience ratings are of lesser importance,  producing Romance is a safe bet, while Sci-Fi and Fantasy movies can be both really well-received and thoroughly disliked by the audience.

### What genres of movies have shown the highest ROI in the last five years?

Though filmmaking is an art form, running a movie production studio is profit-driven, like any other business venture. Hugely successful blockbuster movies are known to produce astronomical ROI (e.g., the very first Star Wars movie in the 70s had a 2500% ROI). In this next phase, the analysis explores the relationship between genres and ROI.

````python
#Create a separate DataFrame where average ROI is grouped by genres
genre_roi = revenue_genre.groupby('genres')['roi'].mean().round()
````

````python
#Organizing and preparing the DataFrame for plotting by choosing the top 15 genres and sort them by highest ROI
x = genre_roi.sort_values(ascending=False).head(15).to_frame()
x.rename(columns={'roi':'avg_roi'}, inplace=True)
````

````python
#Create a bar plot that shows which genres have had the highest ROI in the last five years
ax_four = sns.barplot(x= x.index, y=x.avg_roi, errwidth=0, palette='pastel' )
plt.tight_layout()
ax_four.set_title("Average ROI by Genre for Movies Released (2015-2019)", size = 18)
ax_four.set_xlabel('Movie Genre', size=14)
ax_four.set_ylabel('Average ROI in %', size = 14)
sns.set(rc={'figure.figsize':(16,8)})
````

````python
# To further understand why certain genres have higher ROI than others, let's create a scatter plot
# that explores the relationship between production budget and worldwide gross revenue by genre

ax_five = sns.scatterplot(data = revenue_genre, x='budget', y='ww gross', hue='genres', size='genres',
            sizes=(90,90), palette='colorblind')
sns.set(rc={'figure.figsize':(16,8)})
plt.xlim(0,350000000)
plt.ylim(0,2500000000)
ax_five.set_title("Budget vs Gross: 2015 - 2018", size = 18)
ax_five.set_xlabel('Budget in $', size=14)
ax_five.set_ylabel('Worldwide Gross Revenue in $', size = 14)
ax_five.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), ncol=1)
````

### Conclusion

#### Summary

**Mystery, Horror, Thriller, and Sport** are the only genres that averge over 1000% in ROI. They all tend to cost less than $50M to produce. 

**Biography** movies seem to have the lowest ROI among the 15 genres. Given that they appear in the bottom left corner of the scatter plot, identifying their production cost as comparatively low suggests that that type of movies receive less exposure and probably appeal to a more niche audience that rates them higher. 

**Sci-Fi, Thriller, and Fantasy** films are the ones with the potential of generating the highest ROI if they have a high production budget above $200M. However, on average, Sci-Fi and Fantasy have below 500% ROI, driven by their generally high production budget.

#### Key Takeaways

 - If you are aiming for high risk, high reward type of movie, your best bet is produce a high-budget Sci-Fi/Thriller/Fantasy film.
 - If you want to achieve more with less, focus on producing high-quality Thrillers. Their budgets average below $100M but they are also likely to provide an ROI in the 750-1000% range.


### What genres of movies do the leading 10 production studios create the most?

Being the new kid on the block can be intimidating, especially when you have to compete with industry titans that have dominated the movie markets for decades. Let's take a look at what the ten most successful studios are doing with their projects, movies of which genre they produce the most.  It will then be helpful to explore how these particular genres perform with revenue and ratings within the context of these ten studios only.

````python
#Figure out what are the top 10 production by total gross over the last five years
top_studios = revenue_df.groupby('studio')['ww gross'].sum().sort_values(ascending=False).head(10)
top_studios_list = list(top_studios.index)
print(top_studios_list)
````

````python
#Create a new DataFrame with data for just these top 10 studios
studios_and_movies = revenue_genre[revenue_genre['studio'].isin(top_studios_list)]
````

````python
#Create a count plot showing the number of movies by genre produce by the top 10 studios
ax_six = sns.countplot(x='genres', data=studios_and_movies)
ax_six.set_xticklabels(ax_six.get_xticklabels(), 
                          rotation=30, 
                          horizontalalignment='right')
ax_six.set_title("Count of Movies by Genre Produced by the Top 10 Studios", size = 18)
ax_six.set_xlabel('Movie Genre', size=14)
ax_six.set_ylabel('Count Movies', size = 14)
````

Looking at this breakdown, genres like Action, Adventure, Comedy and Drama seem to be slighly oversaturated.

Based on the takeaways from the previous questions, let's explore how the Sci-Fi, Fantasy, Mystery and Horror movies produced only by the top 10 production studios perform in terms budget vs. gross, and then in terms of user ratings.

````python
#Create a DataFrame sliced by the top 10 studios and including only movies in the four genres.
explore_genres_list = ['Sci-Fi','Fantasy', "Mystery", "Horror"]
explore_genres = revenue_genre[revenue_genre['genres'].isin(explore_genres_list)]
````

````python
explore_genres.head()
````

````python
#Createa a scatter plot to get a closer look at the budget vs gross revenue relationships for the 4 genres
ax_seven = sns.scatterplot(data = explore_genres, x='budget', y='domestic gross', hue='genres', size='genres',
            sizes=(90,90), palette='colorblind')
sns.set(rc={'figure.figsize':(16,8)})
plt.xlim(0,350000000)
plt.ylim(0,1000000000)
ax_seven.set_title("Budget vs Gross: Mystery, Sci-Fi, Fantasy and Horror", size = 18)
ax_seven.set_xlabel('Budget in $', size=14)
ax_seven.set_ylabel('Domestic Gross Revenue in $', size = 14)
ax_seven.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), ncol=1)
````

````python
#Createa a scatter plot to get a closer look at the ratings for the 4 genres

ax_eight = sns.scatterplot(data = explore_genres, x='ww gross', y='rating', hue='genres', size='genres',
            sizes=(90,90), palette='colorblind')
sns.set(rc={'figure.figsize':(16,8)})
ax_eight.set_title("Rating vs Gross: Mystery, Sci-Fi, Fantasy and Horror", size = 18)
ax_eight.set_xlabel('WW Gross in $', size=14)
ax_eight.set_ylabel('User Rating', size = 14)
ax_eight.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), ncol=1)
````

### Conclusion

#### Summary

The top ten movie studios produce the most movies in the genres of **Action, Adventure, Comedy, and Drama** - with over 100 films that can be classified as one or more of these genres for the last five years. 

At the other end of the spectrum, they've produced the fewest **Music/Musical/Documentary/Sports** movies. 

Based on the last takeaways, looking closer at the performance of movies in the **Sci-Fi/Fantasy/Mystery** genres provided additional insight into these genres' potential, all of which are produced at a decent rate but are not oversaturated. 

**Fantasy and Sci-Fi** tend to provide the biggest ROI and can reach really high gross revenue both domestically and globally if they were produced with a higher budget.

High-grossing **Fantasy and Sci-Fi** movies also tend to receive above-average user ratings.


#### Key Takeaways

- Investing in a high-quality Fantasy/Sci-Fi movie will be a great way to launch a movie studio for Microsoft. Not only are these genres not as saturated as others, but they have the highest potential to perform exceptionally well both at the box office and with fans, generating high ROI and favorable user ratings.
